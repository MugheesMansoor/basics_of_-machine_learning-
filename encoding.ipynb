```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#data load 
df=sns.load_dataset('tips')
df.head()
```

```python
df['smoker'].value_counts()
```

```python
#lets do the categorical value change in label encoding  with sk-learn 
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,OrdinalEncoder
le=LabelEncoder()
df['ENcoded_smoker']=le.fit_transform(df['smoker'])
df.head()
```

```python
df['ordinal encoder'].value_counts()
```

```python
oe=OrdinalEncoder(categories=[['Sun','Fri','Sat','Thur']])
df['ordinal encoder']=oe.fit_transform(df[['day']])
df
```
